```python
import pandas as pd
import glob
import os

path = r'../raw_data/cnn_news' # use your path
all_files = glob.glob(path + "/*.csv")

df_list = (pd.read_csv(file, encoding='utf-8-sig') for file in all_files)

combined_df   = pd.concat(df_list, ignore_index=True)
combined_df = combined_df.loc[:,['url','content']]
combined_df
```

```python
combined_df.drop_duplicates(inplace=True)
len(combined_df)
```

```python
import re

content = combined_df['content']
content.head()
len(content)
```

```python
import re

content = combined_df['content']
content = [w[7:] for w in content]


res = []
for sub in content:
        res.append(re.sub('\r\n', '', sub))
res
```

```python
# remove emojis
emoji_pattern = re.compile("["
                u"\U0001F600-\U0001F64F"  # emoticons
                u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                u"\U0001F680-\U0001F6FF"  # transport & map symbols
                u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                u"\U00002500-\U00002BEF"  # chinese char
                u"\U00002702-\U000027B0"
                u"\U00002702-\U000027B0"
                u"\U000024C2-\U0001F251"
                u"\U0001f926-\U0001f937"
                u"\U00010000-\U0010ffff"
                u"\u2640-\u2642"
                u"\u2600-\u2B55"
                u"\u200d"
                u"\u23cf"
                u"\u23e9"
                u"\u231a"
                u"\ufe0f"  # dingbats
                u"\u3030"          
                "]+", flags=re.UNICODE)


remove_emoji = [emoji_pattern.sub(r'',word) for word in res]
remove_emoji
```

```python
import string
remove_punc = [doc.translate(str.maketrans(' ',' ',string.punctuation)) for doc in remove_emoji]
remove_punc
# print(string.punctuation)
```

```python
content_lower=[word.lower() for word in remove_punc]
content_lower
```

```python
remove_singlecomma = [doc.replace("’",' ') for doc in content_lower]
remove_opencomma = [doc.replace("“",' ') for doc in remove_singlecomma]
remove_closecomma = [doc.replace("”",' ') for doc in remove_opencomma]
remove_dash = [doc.replace("–",' ') for doc in remove_closecomma]
remove_extraspaces = [re.sub(' +', ' ', doc) for doc in remove_dash]

remove_extraspaces
```

```python
import re
pattern_order=r'[0-9]'
remove_num=[re.sub(pattern_order,'',doc) for doc in remove_extraspaces]
remove_num
```

```python
tokenize_content = [word.split(' ') for word in remove_num]
# len(tokenize_content)
```

```python
from nltk.corpus import stopwords
stop_list = stopwords.words('english')
remove_stopwords = [[w for w in doc if w not in stop_list] for doc in tokenize_content]
remove_stopwords
```

```python
from nltk.stem.porter import *
stemmer = PorterStemmer()
stemwords = [[stemmer.stem(w) for w in doc] for doc in remove_stopwords]
stemwords
```
