```python
import pandas as pd
import glob
import os

path = r'../raw_data/cna_news' # use your path
all_files = glob.glob(path + "/*.csv")

df_list = (pd.read_csv(file) for file in all_files)
combined_df   = pd.concat(df_list, ignore_index=True)
combined_df = combined_df.loc[:,['url','content']]
combined_df
```

```python
combined_df.drop_duplicates(inplace=True)
len(combined_df)
```

```python
import re

content =combined_df['content']

res = []
for sub in content:
        res.append(re.sub('\r\n', '', sub))
res
```

```python
import string
remove_punc = [doc.translate(str.maketrans(' ',' ',string.punctuation)) for doc in res]
remove_punc
```

```python
content_lower=[word.lower() for word in remove_punc]
content_lower
```

```python
import re
pattern_order=r'[0-9]'
remove_num=[re.sub(pattern_order,'',doc) for doc in content_lower]
remove_num
```

```python
tokenize_content = [word.split(' ') for word in remove_num]
len(tokenize_content)
```

```python
# remove empty words
clean_tokenize_content = [[word for word in sentence if word!=''] for sentence in tokenize_content]
clean_tokenize_content
```

```python
single_tokenize_list=[word for sentence in clean_tokenize_content for word in sentence]
single_tokenize_list
```

```python
from nltk.corpus import stopwords
stop_list = stopwords.words('english')
remove_stopwords = [[w for w in doc if w not in stop_list] for doc in clean_tokenize_content]
remove_stopwords
```

```python
from nltk.stem.porter import *
stemmer = PorterStemmer()
stemwords = [[stemmer.stem(w) for w in doc] for doc in remove_stopwords]
stemwords
```
