```python
import pandas as pd
import glob
import os

path = r'cnn_news' # use your path
all_files = glob.glob(path + "/*.csv")

df_list = (pd.read_csv(file) for file in all_files)
combined_df   = pd.concat(df_list, ignore_index=True)
combined_df = combined_df.loc[:,['url','content']]
combined_df
```

```python
combined_df.drop_duplicates(inplace=True)
len(combined_df)
```

```python
import re

content =combined_df['content']
content=[w[9:] for w in content]

res = []
for sub in content:
        res.append(re.sub('\r\n', '', sub))
res
```

```python
combined_df['clean_content'] = res
combined_df = combined_df.drop('content', axis=1)
```
