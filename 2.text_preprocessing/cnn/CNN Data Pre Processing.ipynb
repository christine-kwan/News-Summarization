Remove Next line in strings

```python
import re

content = df['text']



res = []
for sub in content:
        res.append(re.sub('\r\n', '', sub))
res
```

Remove (CNN) from each string

```python
remove_cnn = [re.sub(r'\(CNN\)\s*', '', text) for text in res]
remove_cnn
```

Remove escape character
- Limitation: Text editor automatically adds in the escape character

```python
remove_esc = [text.replace('\\', '') for text in remove_cnn]
remove_esc
```

Remove emojis from strings

```python
# remove emojis
emoji_pattern = re.compile("["
                u"\U0001F600-\U0001F64F"  # emoticons
                u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                u"\U0001F680-\U0001F6FF"  # transport & map symbols
                u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                u"\U00002500-\U00002BEF"  # chinese char
                u"\U00002702-\U000027B0"
                u"\U00002702-\U000027B0"
                u"\U000024C2-\U0001F251"
                u"\U0001f926-\U0001f937"
                u"\U00010000-\U0010ffff"
                u"\u2640-\u2642"
                u"\u2600-\u2B55"
                u"\u200d"
                u"\u23cf"
                u"\u23e9"
                u"\u231a"
                u"\ufe0f"  # dingbats
                u"\u3030"          
                "]+", flags=re.UNICODE)


remove_emoji = [emoji_pattern.sub(r'',word) for word in remove_esc]
remove_emoji
```

```python
import unicodedata

def removeASCII(text):
    # Normalize the string using Unicode Normalization Form C (NFC)
    normalized_text = unicodedata.normalize('NFC', text)

    # Remove non-ASCII characters from the normalized string
    clean_text = normalized_text.encode('ascii', 'ignore').decode()

    return clean_text

remove_ascii =[removeASCII(word) for word in remove_emoji]
remove_ascii
```

Remove links from string

```python
remove_links=[re.sub(r'http\S+', '', word, flags=re.MULTILINE) for word in remove_ascii]
remove_links
```

Lower case all characters in the string

```python
content_lower=[word.lower() for word in remove_links]
content_lower
```

Translate Non-English to English

```python
from googletrans import Translator, constants
from pprint import pprint
from langdetect import detect
def translate(word):
    translator= Translator()
    if detect(word) != 'en':
        translation = translator.translate(word, dest='en')
        return translation.text
    return word
```

```python
translate_nonEng =[translate(text) for text in content_lower]
translate_nonEng
```
